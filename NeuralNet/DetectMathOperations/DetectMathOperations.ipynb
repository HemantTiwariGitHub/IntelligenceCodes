```python
import tensorflow as tf
```

```python
import pandas as pd
import numpy as np
import os
```

```python
def add(val1, val2):
    return val1+val2

def sub(val1, val2):
    return val1-val2

def mul(val1, val2):
    return val1*val2

def div(val1, val2):
    return val1/val2
    
operations = {0 : add,
           1 : sub,
           2 : mul,
           3 : div,
}

def getVal(val1, val2, op) :
    return operations[op](val1, val2)

def WriteTrainFile():
    MAX_VAL =10000
    SIZE_ARR =1000000
    FILE = "arithmetic.csv"
    COUNT_ARITH_OPS = 4
    
    val1Array = np.random.randint(1,MAX_VAL, size = SIZE_ARR)
    val2Array = np.random.randint(1,MAX_VAL, size = SIZE_ARR)
    opRandom =  np.random.randint(COUNT_ARITH_OPS, size = SIZE_ARR)
    
    print("File is " , FILE)
    if (os.path.isfile(FILE)):
        #print("Removed old " , FILE)
        os.remove(FILE) 
    
    fh = open(FILE, "w")
   
   
  

    desc = "V1,V2,Res,Op"+"\n"
    #print (desc)
    fh.write(desc)
    
    for index in range(0,SIZE_ARR):
        val = getVal(val1Array[index], val2Array[index], opRandom[index])
        line = str(val1Array[index]) + "," + str(val2Array[index]) + "," + str(val) + "," + str(opRandom[index]) +"\n"
        #print (line)
        fh.write(line)
    
    fh.close()
    
WriteTrainFile()
print("train file written")
```

```python
arith = pd.read_csv("arithmetic.csv")
```

```python
arith.head()
```

```python
X = arith[['V1','V2','Res']]
print(X.head())
Y = arith ['Op']
print(Y.head())
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(  X, Y, test_size=0.3, random_state=42)
```

```python
import tensorflow.contrib.learn as learn
```

```python
feature_columns1 = [tf.contrib.layers.real_valued_column("", dimension=3)]
classifier = learn.DNNClassifier(hidden_units=[10, 20, 10], n_classes=4, feature_columns=feature_columns1)
```

```python
classifier.fit(X_train,Y_train,steps=1000, batch_size=10000)
```

```python
predictionOps = list(classifier.predict(X_test, as_iterable=True))
```

```python
print (X_test.head())
print(predictionOps)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print (classification_report(Y_test,predictionOps))
print(confusion_matrix(Y_test,predictionOps))
```
